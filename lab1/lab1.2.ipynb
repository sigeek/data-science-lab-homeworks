# Compute the distance between two points

```python
from math import cos, acos, sin
def distance_coords(lat1, lng1, lat2, lng2):
    deg2rad = lambda x: x * 3.141592 / 180
    lat1, lng1, lat2, lng2 = map(deg2rad, [ lat1, lng1, lat2, lng2 ])
    R = 6378100 # Radius of the Earth, in meters
    return R * acos(sin(lat1) * sin(lat2) + cos(lat1) * cos(lat2) * cos(lng1 - lng2))
```

# Find the closest bike given latitude and longitude

```python
array = obj['network']['stations']
given_lat = 45.074512
given_log = 7.694419
mindist = 1000000
for value in array:
    if(value['free_bikes'] > 0):
        dist = distance_coords(given_lat, given_log, value['latitude'], value['longitude'])
        if dist < mindist:
            mindist = dist
            
print(mindist)
        
        
```
